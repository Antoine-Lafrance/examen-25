# Modèle 1: Par fréquence

je commence par batir un algo de recommendation qui suggère tout simplement l'objet le plus acheté par chaque client

```python
#pour obtenir le nombre de commande de chacun
df_quant = orders_df.groupby(by=["CustomerID","ProductID"])["Quantity"].sum().reset_index()
```

```python
#je peux donc pour un utilisateur donné, lui recommander avec la méthode

def get_rec(customerID, df: pd.DataFrame):
    m = df["CustomerID"] == customerID
    idx = df_quant[m]["Quantity"].max() == df_quant[m]["Quantity"]
    return df_quant[m][idx]["ProductID"].values[0]


assert get_rec("C000",df_quant) == "P044"
```

### Le plus acheté par nom


```python
merged = orders_df.merge(df_customers, on="CustomerID")
merged
```

```python
merged["firstName"] = merged["CustomerName"].apply(lambda x: x.split(" ")[0])
```

```python
tableau_achat_par_nom =  pd.DataFrame(merged.groupby(by=["firstName","ProductID"])["Quantity"].sum())
tableau_achat_par_nom.reset_index().groupby("firstName")[["ProductID","Quantity"]].max()
```

## articles moins vendeurs par région

```python
by_region_min = merged.groupby(by=["Location","ProductID"])["Quantity"].sum().reset_index()

by_region_min.groupby(by=["Location"])[["ProductID","Quantity"]].agg({"ProductID": "first", "Quantity": "min"})
```

composed_transactions
